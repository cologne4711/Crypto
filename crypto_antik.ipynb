### Ersetzen von Buchstaben

Sie können Buchstaben durch andere ersetzen.

**Achtung:** Wenn Sie erst 'a' durch 'b' ersetzen und dann 'b' durch 'c', dann werden am Ende sowohl 'a' als auch 'b' durch 'c' ersetzt sein.

```python
text = "Drei Chinesen mit nem Kontrabass"
text = text.upper()
text = text.replace('A','I').replace('E','I').replace('O','I').replace('U','I') # Hier wird gleich mehrfach ersetzt
print(text)
text = text.replace('I','A')  # Alle I zu A
print(text)
print(text.replace('A','O'))
```

```python
print(5 % 3)
print(26 % 26)
```

## Einschub: Dictionarys

Ein Dictionary ist ähnlich wie eine Liste. Man kann aber anstelle von Indizes auch Schlüsselwörter verwenden. Es ist also so etwas wie ein Telefonbuch.

```python
my_dict = {} # leeres Dictionary
my_dict = {       # Verschiedene Schlüsselwörter mit Werten (gemischte Typen) hinzufügen
    'Peter' : 'Studer',
    'Urs'   : 'Onkel' ,
    'Hallo' : 1234
}
print(my_dict)
print(my_dict['Peter'])
my_dict['Hans'] = 'Meyer' # Schlüsselwort 'Hans' hinzufügen mit Wert 'Meyer'
print(my_dict)
```
